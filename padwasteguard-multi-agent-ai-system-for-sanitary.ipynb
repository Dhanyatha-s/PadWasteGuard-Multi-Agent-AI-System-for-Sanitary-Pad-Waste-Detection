````python
import os
import json
import time
from datetime import datetime
from typing import Dict, List, Optional
import numpy as np
from pathlib import Path

# ADK Imports (Following Day 4 Pattern)
from google.adk.agents import LlmAgent
from google.adk.models.google_llm import Gemini
from google.adk.tools.base_tool import BaseTool
from google.adk.runners import InMemoryRunner
from google.adk.plugins.logging_plugin import LoggingPlugin
from google.adk.tools.agent_tool import AgentTool
from google.adk.tools import google_search

# External libraries
from ultralytics import YOLO
import PIL.Image as Image
import google.generativeai as genai
from google.genai import types
````

````python
class Config:
    """Configuration for autonomous detection"""
    
    # API Keys
    from kaggle_secrets import UserSecretsClient
    user_secrets = UserSecretsClient()
    GOOGLE_API_KEY = user_secrets.get_secret("capstone_project_api")
    
    # Google Custom Search API (Free tier: 100 queries/day)
    # Get from: https://developers.google.com/custom-search/v1/overview
    try:
        GOOGLE_SEARCH_API_KEY = user_secrets.get_secret("GOOGLE_SEARCH_API_KEY")
        GOOGLE_SEARCH_ENGINE_ID = user_secrets.get_secret("GOOGLE_SEARCH_ENGINE_ID")
    except:
        GOOGLE_SEARCH_API_KEY = None
        GOOGLE_SEARCH_ENGINE_ID = None
    
    # Models
    YOLO_MODEL = 'yolov8n.pt'
    GEMINI_MODEL = 'gemini-2.5-flash-lite'
    
    # Detection settings
    CONFIDENCE_THRESHOLD = 0.3
    MAX_IMAGES_TO_ANALYZE = 5  # Number of images to fetch and analyze

# Configure GenAI
genai.configure(api_key=Config.GOOGLE_API_KEY)

retry_config = types.HttpRetryOptions(
    attempts=5,
    exp_base=7,
    initial_delay=1,
    http_status_codes=[429, 500, 503, 504],
)
````

````python
 #==========================================
# AUTONOMOUS DETECTION TOOLS
# ==========================================

class GoogleImageSearchTool(BaseTool):
    """Search Google Images for location-based waste imagery"""
    
    def __init__(self):
        super().__init__(
            name="search_location_images",
            description="Searches Google Images for waste and garbage in a specific location"
        )
        self.api_key = Config.GOOGLE_SEARCH_API_KEY
        self.engine_id = Config.GOOGLE_SEARCH_ENGINE_ID
        self.model = genai.GenerativeModel(Config.GEMINI_MODEL)
    
    async def run(self, location: str, search_terms: List[str] = None) -> Dict:
        """
        Search for images of waste/garbage at a location
        
        Args:
            location: Location to search (e.g., "Marine Beach Chennai")
            search_terms: Additional search terms
            
        Returns:
            Dictionary with image URLs and metadata
        """
        if search_terms is None:
            search_terms = [
                "garbage waste",
                "sanitary pad waste",
                "street waste",
                "debris pollution",
                "waste dump"
            ]
        
        all_images = []
        
        # Method 1: Google Custom Search API with Image Search
        if self.api_key and self.engine_id:
            try:
                for term in search_terms[:3]:  # Limit to avoid quota
                    query = f"{location} {term}"
                    url = "https://www.googleapis.com/customsearch/v1"
                    params = {
                        'key': self.api_key,
                        'cx': self.engine_id,
                        'q': query,
                        'searchType': 'image',
                        'num': 3,  # Get 3 images per search term
                        'imgType': 'photo',
                        'safe': 'off',
                        'fileType': 'jpg,png'
                    }
                    
                    response = requests.get(url, params=params, timeout=10)
                    if response.status_code == 200:
                        data = response.json()
                        for item in data.get('items', []):
                            all_images.append({
                                'url': item.get('link'),
                                'thumbnail': item.get('image', {}).get('thumbnailLink'),
                                'title': item.get('title', 'Untitled'),
                                'source': item.get('displayLink', 'Unknown'),
                                'context': item.get('snippet', ''),
                                'search_query': query
                            })
                
                if all_images:
                    return {
                        'location': location,
                        'images_found': len(all_images),
                        'images': all_images[:Config.MAX_IMAGES_TO_ANALYZE],
                        'method': 'Google Image Search API',
                        'status': 'success'
                    }
                    
            except Exception as e:
                print(f"Google Image Search API failed: {e}")
        
        # Method 2: Fallback - Use Gemini to suggest known waste hotspots
        try:
            prompt = f"""
            For location: {location}
            
            Based on your knowledge, provide information about waste/garbage issues in this area:
            
            1. Known waste hotspots or dumping areas
            2. Common waste management problems
            3. Types of waste typically found (sanitary, plastic, organic, etc.)
            4. Environmental concerns
            5. Recent news or reports about waste in this area
            
            Format as JSON:
            {{
              "waste_hotspots": ["area1", "area2"],
              "common_waste_types": ["type1", "type2"],
              "environmental_concerns": ["concern1", "concern2"],
              "severity_estimate": "low/medium/high/critical",
              "known_issues": ["issue1", "issue2"],
              "simulated_detection": {{
                "likely_sanitary_waste": true/false,
                "confidence": 0-100,
                "reasoning": "why this location may have sanitary waste issues"
              }}
            }}
            """
            
            response = self.model.generate_content(prompt)
            result_text = response.text.strip()
            
            if result_text.startswith('```json'):
                result_text = result_text.split('```json')[1].split('```')[0]
            
            knowledge_data = json.loads(result_text)
            
            return {
                'location': location,
                'images_found': 0,
                'images': [],
                'method': 'Gemini Knowledge Base (No API)',
                'status': 'knowledge_based',
                'knowledge_data': knowledge_data
            }
            
        except Exception as e:
            return {
                'location': location,
                'images_found': 0,
                'images': [],
                'method': 'Fallback',
                'status': 'failed',
                'error': str(e)
            }
````

````python
class ImageDownloadAnalysisTool(BaseTool):
    """Download images from URLs and analyze with YOLO"""
    
    def __init__(self):
        super().__init__(
            name="download_and_analyze_image",
            description="Downloads image from URL and analyzes for waste using YOLO"
        )
        self.yolo = YOLO(Config.YOLO_MODEL)
    
    async def run(self, image_url: str, context: str = "") -> Dict:
        """
        Download and analyze image
        
        Args:
            image_url: URL of image to download
            context: Context about the image
            
        Returns:
            Detection results
        """
        try:
            # Download image
            response = requests.get(image_url, timeout=15, headers={
                'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36'
            })
            
            if response.status_code != 200:
                return {
                    'success': False,
                    'error': f'Failed to download image: HTTP {response.status_code}',
                    'url': image_url
                }
            
            # Load image
            img = Image.open(BytesIO(response.content))
            
            # YOLO detection
            results = self.yolo(img, conf=Config.CONFIDENCE_THRESHOLD)
            
            detections = []
            for r in results:
                boxes = r.boxes
                for box in boxes:
                    cls = int(box.cls[0])
                    conf = float(box.conf[0])
                    label = r.names[cls]
                    
                    detections.append({
                        'class': label,
                        'confidence': conf,
                        'bbox': box.xyxy[0].tolist()
                    })
            
            return {
                'success': True,
                'url': image_url,
                'context': context,
                'detections': detections,
                'detection_count': len(detections),
                'detected_objects': list(set([d['class'] for d in detections])),
                'image_size': img.size
            }
            
        except Exception as e:
            return {
                'success': False,
                'error': str(e),
                'url': image_url
            }
````

````python
class WasteVisionAnalysisTool(BaseTool):
    """Analyze downloaded images for sanitary waste using Gemini Vision"""
    
    def __init__(self):
        super().__init__(
            name="analyze_downloaded_image",
            description="Deep analysis of downloaded images for sanitary waste and environmental hazards"
        )
        self.model = genai.GenerativeModel(Config.GEMINI_MODEL)
    
    async def run(self, image_url: str, yolo_results: Dict, location: str) -> Dict:
        """
        Analyze image with Gemini Vision
        
        Args:
            image_url: URL of the image
            yolo_results: Previous YOLO detection results
            location: Location context
            
        Returns:
            Detailed analysis
        """
        try:
            # Download image for Gemini
            response = requests.get(image_url, timeout=15, headers={
                'User-Agent': 'Mozilla/5.0'
            })
            img = Image.open(BytesIO(response.content))
            
            prompt = f"""
            CRITICAL ENVIRONMENTAL ANALYSIS
            
            Location: {location}
            Image Source: {image_url}
            YOLO Detected: {yolo_results.get('detected_objects', [])}
            
            Analyze this image for:
            
            1. **SANITARY PAD WASTE DETECTION**
               - Is sanitary waste visible? (Yes/No)
               - Confidence level (0-100%)
               - Quantity estimate (items count)
               - Condition (fresh/decomposed/burned)
            
            2. **ENVIRONMENTAL HAZARDS**
               - Water contamination risk
               - Soil pollution risk
               - Air quality impact
               - Wildlife/animal exposure
               - Children accessibility
            
            3. **CALAMITY & DANGER ASSESSMENT**
               - Disease transmission risk (bacterial, viral)
               - Toxic chemical leaching
               - Fire hazard potential
               - Flooding/water contamination if near water bodies
               - Public health emergency level
            
            4. **LOCATION CHARACTERISTICS**
               - Type: residential/commercial/industrial/natural
               - Population exposure: high/medium/low
               - Drainage/water proximity
               - Vegetation/wildlife presence
            
            5. **URGENCY CLASSIFICATION**
               - Immediate danger (0-24 hrs): Critical public health risk
               - Urgent (1-3 days): Significant contamination risk
               - High priority (3-7 days): Growing environmental concern
               - Moderate (1-2 weeks): Manageable with planning
               - Low (>2 weeks): Routine cleanup
            
            Respond in JSON:
            {{
              "sanitary_waste_detected": true/false,
              "detection_confidence": 0-100,
              "waste_quantity": {{
                "estimated_items": number,
                "volume_estimate": "small/medium/large/massive"
              }},
              "environmental_hazards": {{
                "water_contamination_risk": "none/low/medium/high/critical",
                "soil_pollution_risk": "none/low/medium/high/critical",
                "air_quality_impact": "none/low/medium/high/critical",
                "disease_transmission_risk": "none/low/medium/high/critical"
              }},
              "danger_assessment": {{
                "public_health_threat": "none/low/medium/high/critical",
                "children_at_risk": true/false,
                "wildlife_exposure": true/false,
                "toxic_leaching_potential": "none/low/medium/high/critical",
                "overall_danger_level": "safe/caution/dangerous/critical/catastrophic"
              }},
              "calamity_potential": {{
                "disease_outbreak_risk": "none/low/medium/high/imminent",
                "water_supply_contamination": "none/low/medium/high/critical",
                "epidemic_trigger": true/false,
                "environmental_disaster": true/false
              }},
              "urgency": "immediate/urgent/high/moderate/low",
              "urgency_hours": number,
              "location_analysis": {{
                "area_type": "string",
                "population_density": "low/medium/high",
                "sensitive_area": true/false,
                "near_water_body": true/false,
                "near_residential": true/false
              }},
              "detailed_description": "comprehensive description of what you see and risks",
              "immediate_concerns": ["concern1", "concern2", "..."]
            }}
            """
            
            response = self.model.generate_content([prompt, img])
            result_text = response.text.strip()
            
            if result_text.startswith('```json'):
                result_text = result_text.split('```json')[1].split('```')[0]
            
            analysis = json.loads(result_text)
            analysis['image_url'] = image_url
            analysis['yolo_detection_context'] = yolo_results
            
            return analysis
            
        except Exception as e:
            return {
                'error': str(e),
                'sanitary_waste_detected': False,
                'detection_confidence': 0,
                'danger_assessment': {'overall_danger_level': 'unknown'},
                'urgency': 'unknown'
            }
````

````python
# ==========================================
# AGENTS
# ==========================================

# Image Search Agent
image_search_agent = LlmAgent(
    name="image_search_agent",
    model=Gemini(model=Config.GEMINI_MODEL, retry_options=retry_config),
    description="Searches Google Images for waste at specified locations",
    instruction="""
    You search for images of waste and garbage at given locations.
    
    Use search_location_images to find:
    - Recent waste/garbage imagery
    - Sanitary waste images
    - Environmental pollution photos
    - Street waste documentation
    
    Return all found image URLs with context.
    """,
    tools=[GoogleImageSearchTool()]
)

# Image Analysis Agent
image_analysis_agent = LlmAgent(
    name="image_analysis_agent",
    model=Gemini(model=Config.GEMINI_MODEL, retry_options=retry_config),
    description="Downloads and analyzes images with YOLO and Gemini Vision",
    instruction="""
    You download images and perform deep analysis.
    
    For each image:
    1. Use download_and_analyze_image for YOLO detection
    2. Use analyze_downloaded_image for detailed Gemini Vision analysis
    
    Focus on identifying sanitary waste and environmental hazards.
    """,
    tools=[ImageDownloadAnalysisTool(), WasteVisionAnalysisTool()]
)

# Facility Search Agent
facility_agent = LlmAgent(
    name="facility_search_agent",
    model=Gemini(model=Config.GEMINI_MODEL, retry_options=retry_config),
    description="Finds waste management facilities and resources",
    instruction="""
    Search for waste management resources near the location.
    Provide contact information and facility details.
    """,
    tools=[GoogleImageSearchTool()]  # Reuse from previous version
)

# Report Generation Agent
report_agent = LlmAgent(
    name="danger_report_agent",
    model=Gemini(model=Config.GEMINI_MODEL, retry_options=retry_config),
    description="Generates comprehensive danger assessment reports",
    instruction="""
    Generate detailed environmental hazard reports.
    
    Use generate_danger_report to create:
    - Danger assessments
    - Calamity potential analysis
    - Public health risk evaluations
    - Actionable emergency responses
    
    Be thorough and professional.
    """,
    tools=[ComprehensiveReportTool()]
)
````

````python
# ==========================================
# ROOT ORCHESTRATOR
# ==========================================

root_agent = LlmAgent(
    name="autonomous_waste_detector",
    model=Gemini(model=Config.GEMINI_MODEL, retry_options=retry_config),
    instruction="""
    You are an AUTONOMOUS WASTE DETECTION SYSTEM.
    
    When given ONLY a location (no image), you:
    
    1. **SEARCH**: Use image_search_agent to find waste images at that location
    2. **ANALYZE**: Use image_analysis_agent to detect waste in found images
    3. **ASSESS**: Evaluate danger levels and calamity potential
    4. **RESOURCE**: Use facility_search_agent to find cleanup resources
    5. **REPORT**: Use danger_report_agent to generate comprehensive report
    
    Your goal: Autonomous detection and danger assessment based purely on location.
    
    Process ALL found images thoroughly and provide complete danger analysis.
    """,
    tools=[
        AgentTool(agent=image_search_agent),
        AgentTool(agent=image_analysis_agent),
        AgentTool(agent=facility_agent),
        AgentTool(agent=report_agent)
    ]
)
````

````python
# ============================================================
# PadWasteGuard - WORKING FINAL VERSION
# Simplified, Direct Execution - No Awaiting, Immediate Results
# ============================================================

import os
import json
import time
from datetime import datetime
from typing import Dict, List
import numpy as np
from pathlib import Path

# Direct imports - no complex ADK
import google.generativeai as genai
from ultralytics import YOLO
from PIL import Image

# ==========================================
# CONFIGURATION
# ==========================================

class Config:
    from kaggle_secrets import UserSecretsClient
    user_secrets = UserSecretsClient()
    GOOGLE_API_KEY = user_secrets.get_secret("capstone_project_api")
    
    YOLO_MODEL = 'yolov8n.pt'
    GEMINI_MODEL = 'gemini-2.5-flash-lite'
    CONFIDENCE_THRESHOLD = 0.3

# Configure GenAI
genai.configure(api_key=Config.GOOGLE_API_KEY)

# ==========================================
# WORKING AGENTS - DIRECT EXECUTION
# ==========================================

class WorkingVisionAgent:
    """Direct execution - no waiting"""
    
    def __init__(self):
        print("[INIT] Loading YOLO...")
        self.yolo = YOLO(Config.YOLO_MODEL)
        print("[INIT] Loading Gemini...")
        self.gemini = genai.GenerativeModel(Config.GEMINI_MODEL)
        print("[INIT] 笨 Vision Agent Ready")
    
    def detect_waste(self, image_path: str) -> Dict:
        """Execute immediately - return results"""
        
        print(f"\n[YOLO] Analyzing: {Path(image_path).name}")
        
        # YOLO Detection
        results = self.yolo(image_path, conf=Config.CONFIDENCE_THRESHOLD)
        
        detections = []
        for r in results:
            for box in r.boxes:
                cls = int(box.cls[0])
                conf = float(box.conf[0])
                label = r.names[cls]
                detections.append({
                    'object': label,
                    'confidence': f"{conf:.2%}"
                })
        
        print(f"[YOLO] Found {len(detections)} objects")
        
        # Gemini Classification
        print("[GEMINI] Analyzing waste...")
        img = Image.open(image_path)
        
        prompt = f"""
        Analyze this image for sanitary pad waste and environmental hazards.
        
        YOLO detected: {len(detections)} objects
        
        Provide IMMEDIATE analysis in JSON:
        {{
          "pad_waste_detected": true/false,
          "confidence": "85%",
          "waste_description": "brief description",
          "danger_level": "LOW/MEDIUM/HIGH/CRITICAL",
          "environmental_risk": "description",
          "health_hazards": ["hazard1", "hazard2"],
          "urgency": "immediate/24hrs/1week",
          "cleanup_priority": "1-10"
        }}
        """
        
        try:
            response = self.gemini.generate_content([prompt, img])
            text = response.text.strip()
            
            if '```json' in text:
                text = text.split('```json')[1].split('```')[0]
            
            analysis = json.loads(text)
            print(f"[GEMINI] 笨 Analysis complete - Danger: {analysis.get('danger_level', 'UNKNOWN')}")
            
            return {
                'yolo_detections': detections,
                'gemini_analysis': analysis,
                'image_path': image_path
            }
            
        except Exception as e:
            print(f"[ERROR] Gemini failed: {e}")
            return {
                'yolo_detections': detections,
                'gemini_analysis': {'pad_waste_detected': False, 'danger_level': 'UNKNOWN'},
                'image_path': image_path,
                'error': str(e)
            }


class WorkingLocationAgent:
    """Direct location analysis"""
    
    def __init__(self):
        self.gemini = genai.GenerativeModel(Config.GEMINI_MODEL)
        print("[INIT] 笨 Location Agent Ready")
    
    def analyze_location(self, location: str, waste_data: Dict) -> Dict:
        """Execute immediately"""
        
        print(f"\n[LOCATION] Analyzing: {location}")
        
        prompt = f"""
        Location: {location}
        Waste Found: {waste_data.get('gemini_analysis', {}).get('pad_waste_detected', 'Unknown')}
        Danger Level: {waste_data.get('gemini_analysis', {}).get('danger_level', 'Unknown')}
        
        Provide IMMEDIATE location assessment in JSON:
        {{
          "nearest_facilities": [
            {{"name": "facility", "type": "municipal", "distance": "approx"}}
          ],
          "area_risk": "description",
          "population_at_risk": "high/medium/low",
          "water_source_proximity": "yes/no",
          "access_route": "description",
          "google_maps_query": "search query"
        }}
        """
        
        try:
            response = self.gemini.generate_content(prompt)
            text = response.text.strip()
            
            if '```json' in text:
                text = text.split('```json')[1].split('```')[0]
            
            analysis = json.loads(text)
            print("[LOCATION] 笨 Location analysis complete")
            return analysis
            
        except Exception as e:
            print(f"[ERROR] Location analysis failed: {e}")
            return {'error': str(e), 'google_maps_query': f"{location} waste management"}


class WorkingReportAgent:
    """Direct report generation"""
    
    def __init__(self):
        self.gemini = genai.GenerativeModel(Config.GEMINI_MODEL)
        print("[INIT] 笨 Report Agent Ready")
    
    def generate_report(self, detection: Dict, location_data: Dict, location_name: str) -> str:
        """Execute immediately"""
        
        print("\n[REPORT] Generating comprehensive report...")
        
        prompt = f"""
        Generate IMMEDIATE action report:
        
        DETECTION DATA:
        {json.dumps(detection.get('gemini_analysis', {}), indent=2)}
        
        LOCATION DATA:
        {json.dumps(location_data, indent=2)}
        
        Create professional report:
        
        # URGENT WASTE DETECTION REPORT
        Location: {location_name}
        Time: {datetime.now().strftime('%Y-%m-%d %H:%M')}
        
        ## EXECUTIVE SUMMARY
        [Brief critical findings]
        
        ## DETECTION DETAILS
        - Pad Waste: [Yes/No]
        - Danger Level: [Level]
        - Confidence: [%]
        
        ## DANGER ASSESSMENT
        - Environmental Risk: [Description]
        - Health Hazards: [List]
        - Population at Risk: [Assessment]
        
        ## IMMEDIATE ACTIONS REQUIRED
        1. [Action 1]
        2. [Action 2]
        3. [Action 3]
        
        ## AVAILABLE RESOURCES
        - Facilities: [List]
        - Contact: [Info]
        - Access: [Route]
        
        ## GOOGLE MAPS
        Search: {location_data.get('google_maps_query', 'N/A')}
        
        Use bullet points, be direct, be actionable.
        """
        
        try:
            response = self.gemini.generate_content(prompt)
            print("[REPORT] 笨 Report generated")
            return response.text
            
        except Exception as e:
            print(f"[ERROR] Report generation failed: {e}")
            return f"# ERROR IN REPORT\n\n{str(e)}\n\nRaw Data:\n{json.dumps(detection, indent=2)}"


# ==========================================
# MAIN ORCHESTRATOR - DIRECT EXECUTION
# ==========================================

class PadWasteGuardWorking:
    """Direct execution - no async, no waiting"""
    
    def __init__(self):
        print("\n" + "="*70)
        print("泅 INITIALIZING PADWASTEGUARD")
        print("="*70)
        
        self.vision = WorkingVisionAgent()
        self.location = WorkingLocationAgent()
        self.reporter = WorkingReportAgent()
        
        print("\n笨 ALL SYSTEMS READY")
        print("="*70)
    
    def analyze_single_image(self, image_path: str, location: str = "Chennai, India"):
        """Execute complete analysis - IMMEDIATE RESULTS"""
        
        print(f"\n{'='*70}")
        print(f"沒 TARGET: {location}")
        print(f"泱ｼｸ  IMAGE: {Path(image_path).name}")
        print('='*70)
        
        # Step 1: Vision Detection
        print("\n[PHASE 1] WASTE DETECTION")
        detection = self.vision.detect_waste(image_path)
        
        # Step 2: Location Analysis
        print("\n[PHASE 2] LOCATION ANALYSIS")
        location_data = self.location.analyze_location(location, detection)
        
        # Step 3: Report Generation
        print("\n[PHASE 3] REPORT GENERATION")
        report = self.reporter.generate_report(detection, location_data, location)
        
        print("\n" + "="*70)
        print("笨 ANALYSIS COMPLETE")
        print("="*70)
        
        return {
            'detection': detection,
            'location': location_data,
            'report': report
        }
    
    def analyze_batch(self, image_paths: List[str], location: str = "Chennai, India"):
        """Process multiple images"""
        
        results = []
        
        for img_path in image_paths:
            try:
                result = self.analyze_single_image(img_path, location)
                results.append(result)
            except Exception as e:
                print(f"[ERROR] Failed to process {img_path}: {e}")
                results.append({'error': str(e), 'image': img_path})
        
        return results


# ==========================================
# DEMO - IMMEDIATE EXECUTION
# ==========================================

print("\n" + "="*70)
print("泅 PADWASTEGUARD - WORKING FINAL VERSION")
print("="*70)
print("\n汳｡ FEATURES:")
print("   笨 Immediate YOLO Detection")
print("   笨 Instant Gemini Analysis")
print("   笨 Direct Location Assessment")
print("   笨 On-the-Spot Report Generation")
print("   笨 NO WAITING - IMMEDIATE RESULTS")
print("="*70)

# Initialize system
guard = PadWasteGuardWorking()

# Run analysis
test_image = "/kaggle/input/sanitary-pad-waste-detection-dataset/Waste_dataset/image_1.jpg"
test_location = "Chennai, Tamil Nadu, India"

result = guard.analyze_single_image(test_image, test_location)

# Display results
print("\n" + "="*70)
print("沒 FINAL COMPREHENSIVE REPORT")
print("="*70)
print(result['report'])

print("\n" + "="*70)
print("沒 DETECTION SUMMARY")
print("="*70)
print(json.dumps({
    'yolo_objects': len(result['detection']['yolo_detections']),
    'pad_waste': result['detection']['gemini_analysis'].get('pad_waste_detected'),
    'danger_level': result['detection']['gemini_analysis'].get('danger_level'),
    'urgency': result['detection']['gemini_analysis'].get('urgency')
}, indent=2))

print("\n沁 PROCESSING COMPLETE - ALL RESULTS DELIVERED IMMEDIATELY")
````
